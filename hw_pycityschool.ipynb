```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# Identify incomplete rows
school_data_complete.count()
```

```python
#calculate total school number by using unique function to get list of total school then use len to calculate school number
total_schools = len(school_data_complete["school_name"].unique())
total_schools
```

```python
#calculate total students number by calculating student ID 
total_students = school_data_complete["Student ID"].count()
total_students
```

```python
#Calculate the total number of students 2
#total_budget = grouped_total_schools["size"].sum()
total_budget = school_data_complete["budget"].unique().sum()
total_budget
```

```python
#Calculate the average math score
the_average_math_score = school_data_complete["math_score"].mean()
the_average_math_score
```

```python
#Calculate the average reading score
the_average_reading_score = school_data_complete["reading_score"].mean()
the_average_reading_score
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
math_pass = school_data_complete.loc[school_data_complete["math_score"] >= 70, :]
students_number_math = math_pass["Student ID"].count()
students_number_math
```

```python
#Calculate the percentage of students with a passing math score
the_percentage_of_students_with_a_passing_math = (students_number_math / total_students)*100
the_percentage_of_students_with_a_passing_math
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
reading_pass = school_data_complete.loc[school_data_complete["reading_score"] >= 70, :]
student_number_reading = reading_pass["Student ID"].count()
student_number_reading
```

```python
#Calculate the percentage of students with a passing math score
the_percentage_of_students_with_a_passing_reading = (student_number_reading / total_students)*100
the_percentage_of_students_with_a_passing_reading 
```

```python
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

the_overall_passing_rate = (the_percentage_of_students_with_a_passing_reading + the_percentage_of_students_with_a_passing_math )/2
the_overall_passing_rate
```

```python

Summary_1 = pd.DataFrame({"Total Schools": [total_schools],
                          "Total Students": [total_students],
                          "Total Budget": [total_budget],
                          "Average Math Score": [the_average_math_score],
                          "Average Reading Score": [the_average_reading_score],
                          "% Passing Math": [the_percentage_of_students_with_a_passing_math],
                          "% Passing Reading": [the_percentage_of_students_with_a_passing_reading],
                          "% Overall Passing Rate": [the_overall_passing_rate]})
Summary_1
```

```python
# Use Map to format all the columns
Summary_1["Total Students"] = Summary_1["Total Students"].map("{:,}".format)
Summary_1["Total Budget"] = Summary_1["Total Budget"].map("${:,}".format)
Summary_1["Average Math Score"] = Summary_1["Average Math Score"].map("{:.6f}".format)
Summary_1["Average Reading Score"] = Summary_1["Average Reading Score"].map("${:.6f}".format)
Summary_1["% Passing Math"] = Summary_1["% Passing Math"].map("{:.6f}".format)
Summary_1["% Passing Reading"] = Summary_1["% Passing Reading"].map("{:.6f}".format)
Summary_1["% Overall Passing Rate"] = Summary_1["% Overall Passing Rate"].map("{:.6f}".format)
Summary_1
```
